```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../datos/union_netflix.csv")
df.sample(1)
```

```python
df.columns
```

#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
df_ejercicio1 = df.groupby("type")["title"].count()
df_ejercicio1 = df_ejercicio1.reset_index()
df_ejercicio1
```

#### Ejercicio 2: ¿Cuál es el promedio de puntuación de las películas agrupadas por clasificación de audiencia ('rating')? Interpreta los resultados, ¿hay algo que te llame la atención? Explica por qué. 

```python
df.sample(1)
```

```python
df_ejercicio2 = np.round(df.groupby("rating")["IMDB Score"].mean(),2)
df_ejercicio2 = df_ejercicio2.reset_index()
df_ejercicio2.sort_values(by="IMDB Score", ascending=False)
```

#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
df_ejercicio3 = df.groupby("country")["title"].count()
df_ejercicio3 = df_ejercicio3.reset_index()
df_ejercicio3.sort_values(by="title", ascending= False)
```

#### Ejercicio 4: ¿Cuál es la calificación promedio en IMDb de los títulos agrupados por tipo ('Movie' o 'TV Show')?

```python
df.columns
```

```python
df_ejercicio4 = np.round(df.groupby("type")["IMDB Score"].mean(),2)
df_ejercicio4 = df_ejercicio4.reset_index()
df_ejercicio4
```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
df.head(1)
```

```python
df['date_added'] = df['date_added'].str.strip()

df["date_added"] = pd.to_datetime(df["date_added"])

df["year_added"] = df["date_added"].dt.year
```

```python
df.head(1)
```

```python
df.info()
```

```python
df_ejercicio5 = df.groupby("year_added")["Title"].count()
df_ejercicio5 = df_ejercicio5.reset_index()
df_ejercicio5.sort_values(by="Title", ascending= False)
```

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
año_2020 = df["year_added"] == 2020

df_ejercicio6 = df[año_2020].groupby("country")["Title"].count()
df_ejercicio6 = df_ejercicio6.reset_index()
df_ejercicio6.sort_values(by="Title", ascending=False).head(1)
```

#### Ejercicio 7:¿Cuál es el promedio de puntuación en IMDb de los títulos agrupados por país?

```python
df_ejercicio7 = df.groupby("country")["IMDB Score"].mean()
df_ejercicio7 = df_ejercicio7.reset_index()
df_ejercicio7.sort_values(by="IMDB Score", ascending=False).dropna()
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?

```python
df_ejercicio8 = df.groupby(["year_added", "type"])["Title"].count()
df_ejercicio8 = df_ejercicio8.reset_index()
df_ejercicio8.sort_values(by="Title", ascending=False)
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?

```python
df.head(1)
```

```python
df.columns
```

```python
tvma_titles = df["rating"] == "TV-MA"

df_ejercicio9 = df[tvma_titles].groupby("country")["rating"].count()
df_ejercicio9 = df_ejercicio9.reset_index()
pais_rating = pd.DataFrame(df_ejercicio9.max())
pais_rating
```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
df.sample(1)
```

```python
df.info()
```

```python
df["duracion_final"] = df["duration"].str.extract('(\d+)').astype(float)
```

```python
df_ejercicio10 = df.groupby("country")["duracion_final"].mean() 
df_ejercicio10 = df_ejercicio10.reset_index()
df_ejercicio10.sort_values(by="duracion_final", ascending=False).dropna()
```

#### Ejercicio 11: ¿Cuál es el número total de títulos por país y clasificación de audiencia ('rating')?

```python
df_ejercicio11 = df.groupby(["country","rating"])["Title"].count()
df_ejercicio11 = df_ejercicio11.reset_index()
df_ejercicio11.sort_values(by="Title", ascending=False)
```

#### Ejercicio 12: ¿Cuántos títulos fueron lanzados antes del año 2000 agrupados por país?

```python
before_2020 = df["year_added"] < 2000

df_ejercicio12 = df[before_2020].groupby("country")["Title"].count()
df_ejercicio12 = df_ejercicio12.reset_index()
df_ejercicio12.sort_values(by="Title", ascending=False)
```

#### Ejercicio 13: ¿Cuál es la puntuación promedio en IMDb de los títulos lanzados en 2021 agrupados por país?

```python
titles_2021 = df["year_added"] == 2021

df_ejercicio13 = df[titles_2021].groupby("country")["IMDB Score"].mean()
df_ejercicio13 = df_ejercicio13.reset_index()
np.round(df_ejercicio13.sort_values(by="IMDB Score", ascending=False).dropna(),2)
```

#### Ejercicio 14: ¿Cuántos títulos fueron dirigidos por cada director? Ordena los resultados de mayor a menor

```python
df_ejercicio14 = df.groupby("director")["Title"].count()
df_ejercicio14 = df_ejercicio14.reset_index()
df_ejercicio14.sort_values(by="Title", ascending=False)
```
